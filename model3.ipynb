```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy 
%matplotlib inline
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
import os
for dirname, _, filenames in os.walk('C:\\Users\\siddh\\OneDrive\\Desktop\\sem project\\tomato_data\\tomato'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        'C:\\Users\\siddh\\OneDrive\\Desktop\\sem project\\tomato_data\\tomato\\train',
        target_size=(150, 150),
        batch_size=32,
        class_mode='categorical')
```

```python
test_generator = test_datagen.flow_from_directory(
        'C:\\Users\\siddh\\OneDrive\\Desktop\\sem project\\tomato_data\\tomato\\val',
        target_size=(150, 150),
        batch_size=8,
        class_mode='categorical')
```

```python
# Third model

cnn3 = tf.keras.models.Sequential()       

#Convolution layer 1
cnn3.add(tf.keras.layers.Conv2D(filters=32,kernel_size = 3, activation='relu',input_shape=[150, 150,3]))

# Pooling 1
cnn3.add(tf.keras.layers.MaxPool2D(pool_size=1, strides=1))

#Convolution layer 2
cnn3.add(tf.keras.layers.Conv2D(filters=32,kernel_size = 3, activation='relu',input_shape=[150, 150,3]))

# Pooling 2
cnn3.add(tf.keras.layers.MaxPool2D(pool_size=1, strides=1))

# Flattening
cnn3.add(tf.keras.layers.Flatten())

#Full Conncetion
cnn3.add(tf.keras.layers.Dense(units=128,activation='relu'))

#Output Layer
cnn3.add(tf.keras.layers.Dense(units=10,activation='softmax'))

#Compiling
cnn3.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

# Summary of the model
cnn3.summary()

# Fit
temp = cnn3.fit(x = train_generator, validation_data=test_generator,steps_per_epoch=125,validation_steps=125, epochs=25)
```

```python
cnn3.save('third_model.h5')
```
